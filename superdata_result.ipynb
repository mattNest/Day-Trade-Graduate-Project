```python
import pandas as pd
import numpy as np
```

```python
superdata = pd.read_csv('super_allinformation_20130106_20150106.csv', index_col='day', parse_dates=True)
```

```python
superdata.head()
```

```python
superdata['daily_return_volatility'] = 0
```

```python
for i in range(len(superdata.index)-3):
    if(superdata['return'].iloc[i]==0):
        superdata['daily_return_volatility'].iloc[i] = 0
    else:
        superdata['daily_return_volatility'].iloc[i] = (superdata['return'].iloc[i+3] - superdata['return'].iloc[i])/superdata['return'].iloc[i]
        
```

```python
superdata = superdata.dropna() # drop null value
```

```python
superdata
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
xfit = superdata['trade']
yfit = superdata['daily_return_volatility']
```

```python
xfit = np.array(xfit).reshape(-1,1)
```

```python
lin_reg.fit(xfit,yfit)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```
